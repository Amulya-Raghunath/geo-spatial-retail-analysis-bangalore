```python
import json
import pandas as pd
with open("osm_shops_bengaluru.geojson", encoding="utf-8") as f:
    data=json.load(f)
features=data["features"]
features
```

```python
rows = []

for feature in features:
    name = feature["properties"].get("name")
    shop = feature["properties"].get("shop")
    lon, lat = feature["geometry"]["coordinates"]

    rows.append([name, shop, lat, lon])

df = pd.DataFrame(rows, columns=["name","shop","latitude","longitude"])
df.head()
```

```python
df['shop'].value_counts()
```

```python
df.to_csv("osm_shops_bengaluru.csv", index=False)
```
